```python
from transformers import VisionEncoderDecoderModel, TrOCRProcessor

# Load processor and model
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-handwritten")

# ✅ Set decoder start token (very important for training)
model.config.decoder_start_token_id = processor.tokenizer.cls_token_id

# ✅ Set padding token ID to avoid NoneType issues
model.config.pad_token_id = processor.tokenizer.pad_token_id
```

```python
from torchvision.transforms import Compose, Resize, ToTensor
from PIL import Image
import torch

# Image transforms
transform = Compose([
    Resize((384, 384)),
    ToTensor()
])

# Function to preprocess one item
def preprocess_sample(example):
    image = example["image"]
    label = example["label"]

    pixel_values = transform(image)
    labels = processor.tokenizer(label, padding="max_length", truncation=True, max_length=64).input_ids

    return {
        "pixel_values": pixel_values,
        "labels": labels
    }
```

```python
iam_data = [{"image": ex["image"].convert("RGB"), "label": ex["text"]} for ex in iam_dataset.select(range(100))]
```

```python
from PIL import Image
import os

imgur_dir = "/content/IMGUR5K-Handwriting-Dataset/IMGUR5K-Handwriting-Dataset/images"  # Update with actual path

imgur_data = []
for fname in os.listdir(imgur_dir)[:100]:  # Sample 100
    if fname.endswith(('.png', '.jpg')):
        label = os.path.splitext(fname)[0].split("_")[0]  # Extract word from filename
        img_path = os.path.join(imgur_dir, fname)

        try:
            image = Image.open(img_path).convert("RGB")
            imgur_data.append({
                "image": image,
                "label": label
            })
        except Exception as e:
            print(f"Error loading {fname}: {e}")
```

```python
all_data = iam_data + imgur_data # + synthetic_data
```

```python
from datasets import Dataset

# Convert to Hugging Face dataset
dataset = Dataset.from_list(all_data)

# Apply preprocessing
processed_dataset = dataset.map(preprocess_sample, remove_columns=["label"])
```

```python
# Split 90% train, 10% eval
dataset_split = processed_dataset.train_test_split(test_size=0.1)
train_dataset = dataset_split["train"]
eval_dataset = dataset_split["test"]

print(f"Train: {len(train_dataset)}, Eval: {len(eval_dataset)}")
```

```python
from transformers import Seq2SeqTrainingArguments

training_args = Seq2SeqTrainingArguments(
    output_dir="./trocr-finetuned",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=10,
    per_device_eval_batch_size=2,
    predict_with_generate=True,
    num_train_epochs=10,
    logging_dir="./logs",
    logging_steps=10,
    save_total_limit=2,
    fp16=True  # Use GPU mixed precision if available
)
```

```python
from transformers import Seq2SeqTrainer, default_data_collator

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=processor.tokenizer,
    data_collator=default_data_collator
)
```

```python
trainer.train()
```

```python
ground_truths = []
predictions = []

# Evaluate on a few samples
for sample in eval_dataset.select(range(10)):
    pixel_values = torch.tensor(sample["pixel_values"]).unsqueeze(0).to(model.device)
    generated_ids = model.generate(pixel_values)

    pred = processor.tokenizer.decode(generated_ids[0], skip_special_tokens=True)
    true = processor.tokenizer.decode(sample["labels"], skip_special_tokens=True)

    predictions.append(pred)
    ground_truths.append(true)

    print(f"🔹 GT: {true}\n🔸 PR: {pred}\n")
```

```python
from jiwer import cer, wer

cer_score = cer(ground_truths, predictions)
wer_score = wer(ground_truths, predictions)

print(f"🧠 Character Error Rate (CER): {cer_score:.2%}")
print(f"🧠 Word Error Rate (WER): {wer_score:.2%}")
```

```python
from jiwer import cer, wer
import pandas as pd
import torch

def evaluate_model(model, processor, dataset, num_samples=10, save_csv=False, csv_path="predictions.csv"):
    ground_truths = []
    predictions = []
    rows = []

    print(f"Evaluating on {num_samples} samples...\n")

    for sample in dataset.select(range(num_samples)):
        pixel_values = torch.tensor(sample["pixel_values"]).unsqueeze(0).to(model.device)

        # Generate prediction
        generated_ids = model.generate(pixel_values)
        pred = processor.tokenizer.decode(generated_ids[0], skip_special_tokens=True)
        true = processor.tokenizer.decode(sample["labels"], skip_special_tokens=True)

        ground_truths.append(true)
        predictions.append(pred)
        rows.append({"Ground Truth": true, "Prediction": pred})

        print(f"✅ GT: {true}\n🎯 PR: {pred}\n")

    # Compute metrics
    cer_score = cer(ground_truths, predictions)
    wer_score = wer(ground_truths, predictions)

    print("📊 Evaluation Results:")
    print(f"🧠 Character Error Rate (CER): {cer_score:.2%}")
    print(f"🧠 Word Error Rate (WER): {wer_score:.2%}")

    if save_csv:
        df = pd.DataFrame(rows)
        df.to_csv(csv_path, index=False)
        print(f"📄 Predictions saved to {csv_path}")

    return cer_score, wer_score
```

```python
cer_score, wer_score = evaluate_model(model, processor, eval_dataset, num_samples=10, save_csv=True)
```
